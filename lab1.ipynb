```python
import pandas as pd
df = pd.read_csv('/content/reviews_data.csv')
```

```python
df.head(5)
```

```python
# finding missing data
# only at image links
# result_df = df[ df['Image_Links'] ==  'No Images' ]
# print(result_df)
null_records = df[df.isnull().any(axis=1)]

# Print records with null values
print(null_records)
```

```python
df_encoded = pd.get_dummies(df, columns=['name'], prefix=['name'])

# print(df_encoded)
df_encoded.head(5)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoders = {}
for col in ['name', 'location', 'Review']:
    label_encoders[col] = LabelEncoder()
    df[col + '_encoded'] = label_encoders[col].fit_transform(df[col])

df.head()
```

```python
from sklearn.model_selection import train_test_split
# Define your features (X) and target variable (y)
X = df.drop('Rating', axis=1)  # Replace 'target_column_name' with your target variable's name
y = df['Rating']  # Replace 'target_column_name' with your target variable's name

# Split the dataset into a training set and a test set (typically 70-80% for training, and 20-30% for testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(X_train)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Fit and transform the scaler on your feature data (X)
numeric_columns = ['Rating']
X = df[numeric_columns]

# Apply feature scaling using StandardScaler

X_scaled = scaler.fit_transform(X)

# Replace the original numeric columns with the scaled values
df[numeric_columns] = X_scaled

# Now, your DataFrame 'df' contains scaled numeric columns.
print(df)
```
